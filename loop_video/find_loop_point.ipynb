```python
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

def find_most_similar_frames(video_path):
    cap = cv2.VideoCapture(video_path)
    frames = []
    
    # 모든 프레임을 리스트에 저장
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY))  # 흑백 변환하여 저장
    
    cap.release()
    total_frames = len(frames)
    best_pair = (0, 0)
    best_score = -1  # SSIM은 0~1 값, 1에 가까울수록 유사함

    # 모든 프레임을 서로 비교 (중복 비교 방지)
    for i in range(total_frames):
        for j in range(i + 1, total_frames):  # 자기 자신 비교 방지
            score, _ = ssim(frames[i], frames[j], full=True)
            if score > best_score:
                best_score = score
                best_pair = (i, j)

    print(f"가장 유사한 프레임 번호: {best_pair[0]} & {best_pair[1]}, SSIM 점수: {best_score}")
    return best_pair
```

```python
# 사용
# 걸린 시간 : 
your_video = 'Snow Falling - 4K Green screen FREE DOWNLOAD'
similar_frames = find_most_similar_frames(f"{your_video}.mp4")
print(f"루프 후보 프레임: {similar_frames[0]} <-> {similar_frames[1]}")
```
