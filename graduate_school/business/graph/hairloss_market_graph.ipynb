# 기본 세팅

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
# 'subprocess'는 Windows 폰트 설치에 필요 없으므로 제거합니다.
# import subprocess 

# --- 1. 한글 폰트 설정 (Windows 11 환경) ---
# [수정] 알려주신 코드를 적용하여 사용 가능한 폰트를 자동으로 찾습니다.
# 한글 폰트 설정 함수
def setup_korean_font():
    # from matplotlib import font_manager, rcParams # 상단에서 fm으로 import 하므로 중복 제거
    candidates = ["AppleGothic","Malgun Gothic","NanumGothic","Noto Sans CJK KR","Noto Sans KR","NanumBarunGothic"]
    available = set(f.name for f in fm.fontManager.ttflist)
    chosen = None
    for c in candidates:
        if c in available:
            chosen = c; break
    if chosen:
        plt.rcParams["font.family"] = chosen # rcParams 대신 plt.rcParams 사용
    plt.rcParams["axes.unicode_minus"] = False
    return chosen

# [수정] 함수를 호출하고, 찾은 폰트 이름을 변수에 저장합니다.
chosen_font = setup_korean_font()
# ------------------------------------
```

# 해외 탈모 케어 시장 규모 추정

```python
# Data (이전과 동일)
data = {
    '연도': [2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030],
    '시장 규모(억 달러)': [121.62, 144.58, 159.55, 176.03, 194.1, 210.98, 227.86, 246.82, 267.02, 288.59, 311.55],
    '데이터 유형': ['실제 데이터', '실제 데이터', '추정 데이터', '추정 데이터', '추정 데이터', '실제 데이터', '추정 데이터', '추정 데이터', '추정 데이터', '추정 데이터', '추정 데이터']
}
df = pd.DataFrame(data)

# --- 2. Seaborn으로 그래프 그리기 ---

# Seaborn 테마 설정 (스타일, 폰트)
# [수정] "NanumGothic"을 하드코딩하는 대신, 위에서 찾은 폰트(chosen_font)를 사용합니다.
sns.set_theme(style="darkgrid", font=chosen_font)

# 그래프 크기 설정
plt.figure(figsize=(14, 8))

# Seaborn barplot 생성
# 'hue' 옵션을 사용하면 '데이터 유형'에 따라 자동으로 색상을 구분하고 범례를 생성합니다.
ax = sns.barplot(
    data=df,
    x='연도',
    y='시장 규모(억 달러)',
    hue='데이터 유형',
    palette='viridis' # Seaborn의 'viridis' 색상 팔레트 사용
)

# 막대 위에 값 표시 (bar_label 사용)
# hue를 사용했으므로, container가 두 개(실제, 추정)로 나뉩니다.
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', fontsize=11, padding=3)

# 제목 및 레이블 설정
ax.set_title('탈모 치료 해외 시장 규모 추정', fontsize=18, pad=15)
ax.set_xlabel('연도', fontsize=14)
ax.set_ylabel('시장 규모 (억 달러)', fontsize=14)

# Y축 범위 조정 (레이블이 잘리지 않도록)
ax.set_ylim(0, df['시장 규모(억 달러)'].max() * 1.15)

# 범례 위치 조정
ax.legend(loc='upper left', fontsize=12)

# 저장
plt.tight_layout()
# plt.savefig('seaborn_darkgrid_chart.png')

print("seaborn_darkgrid_chart.png saved successfully.")
```

```python
# Data (이전과 동일)
data = {
    '연도': [2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030],
    '시장 규모(억 달러)': [121.62, 144.58, 159.55, 176.03, 194.1, 210.98, 227.86, 246.82, 267.02, 288.59, 311.55],
    '데이터 유형': ['실제 데이터', '실제 데이터', '추정 데이터', '추정 데이터', '추정 데이터', '실제 데이터', '추정 데이터', '추정 데이터', '추정 데이터', '추정 데이터', '추정 데이터']
}
df = pd.DataFrame(data)

# --- 2. Seaborn과 Matplotlib으로 그래프 그리기 (고급 다크 테마) ---

# Seaborn 테마 설정 - 'darkgrid'는 배경이 검정에 가깝지만, 더 세밀한 조정은 Matplotlib으로 합니다.
sns.set_theme(style="whitegrid", font=chosen_font) # whitegrid를 기본으로 하고 수동으로 색상 변경
plt.rcParams['axes.facecolor'] = '#1A1A1A' # 플롯 영역의 배경색을 아주 어두운 회색으로 설정
plt.rcParams['figure.facecolor'] = '#1A1A1A' # 전체 그림의 배경색을 동일하게 설정
plt.rcParams['text.color'] = '#E0E0E0'     # 텍스트 색상을 밝은 회색으로 설정
plt.rcParams['axes.labelcolor'] = '#E0E0E0' # 축 레이블 색상
plt.rcParams['xtick.color'] = '#E0E0E0'    # X축 틱 색상
plt.rcParams['ytick.color'] = '#E0E0E0'    # Y축 틱 색상
plt.rcParams['grid.color'] = '#3A3A3A'     # 그리드 색상 (더 어둡게)
plt.rcParams['grid.linestyle'] = '--'      # 그리드 스타일 (점선)
plt.rcParams['legend.facecolor'] = '#2A2A2A' # 범례 배경색
plt.rcParams['legend.edgecolor'] = '#4A4A4A' # 범례 테두리색
plt.rcParams['axes.edgecolor'] = '#3A3A3A' # 축 테두리색

plt.figure(figsize=(15, 8))

# 색상 팔레트: 깊이감 있는 웜-톤 또는 쿨-톤 그라데이션
# 'crest'나 'rocket' 같은 Seaborn 팔레트는 어두운 배경에 잘 어울립니다.
# 여기서는 'rocket'을 사용하여 짙은 주황색 계열 그라데이션을 적용합니다.
palette = sns.color_palette("rocket", n_colors=len(df['연도']))

# barplot 생성
ax = sns.barplot(
    data=df,
    x='연도',
    y='시장 규모(억 달러)',
    hue='데이터 유형', # 데이터 유형별로 색상을 구분
    palette={'실제 데이터': palette[len(df['연도']) // 3], '추정 데이터': palette[len(df['연도']) // 3 * 2]}, # 실제/추정 데이터에 특정 색상 부여
    # 또는 '연도'에 따른 그라데이션을 사용하고 싶다면, hue를 제거하고 아래 bar_colors를 직접 매핑
    # palette=palette # hue를 사용하지 않고 연도별로 그라데이션을 적용할 경우
    edgecolor='#3A3A3A' # 막대 테두리 색상
)

# 막대 위에 값 표시
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', fontsize=12, padding=5, color='#E0E0E0', weight='bold')

# 제목 및 레이블
ax.set_title('탈모 치료 해외 시장 규모 추정', fontsize=22, pad=25, color='#E0E0E0', weight='bold')
ax.set_xlabel('연도', fontsize=16, labelpad=15, color='#E0E0E0')
ax.set_ylabel('시장 규모 (억 달러)', fontsize=16, labelpad=15, color='#E0E0E0')

# Y축 범위 조정
ax.set_ylim(0, df['시장 규모(억 달러)'].max() * 1.2)

# 축 스파인 (테두리) 설정
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#3A3A3A')
ax.spines['bottom'].set_color('#3A3A3A')

# 틱 파라미터
ax.tick_params(axis='x', length=0, labelsize=12) # x축 틱 길이 제거
ax.tick_params(axis='y', length=4, labelsize=12) # y축 틱 길이 조정

# 범례 설정 (배경과 테두리 색상 설정)
ax.legend(loc='upper left', fontsize=13, frameon=True, facecolor='#2A2A2A', edgecolor='#4A4A4A', labelcolor='#E0E0E0')

# 레이아웃 조정 및 저장
plt.tight_layout()
# plt.savefig('seaborn_dark_theme_pro_chart.png', dpi=300) # 고해상도로 저장

print("seaborn_dark_theme_pro_chart.png saved successfully.")
```
