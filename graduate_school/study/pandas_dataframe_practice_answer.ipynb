## 3. 새로운 컬럼 추가
- Q3-1. temperature_C가 37도 이상인 샘플은 High를, 37이하인 샘플은 Normal인 temp_status컬럼을 만들어주세요
- Q3-2. ph_level이 7미만이면 Acidic, 7초과이면 Basic, 7이면 Neutral인 acid_base컬럼을 만들어주세요.
- Q3-3. score 컬럼을 만들어주세요.(벡터 연산)
    - score = temperature_C * ph_level

```python
# Q3-1. temperature_C가 37도 이상인 샘플은 High를, 37이하인 샘플은 Normal인 temp_status컬럼을 만들어주세요

df['temp_status'] = np.where(df['temperature_C'] >= 37.0, 'High', 'Normal')
df
```

```python
temp = df.copy()
temp
```

```python
# Q3-2. ph_level이 7미만이면 Acidic, 7초과이면 Basic, 7이면 Neutral인 acid_base컬럼을 만들어주세요.

df['acid_base'] = df['ph_level'].apply(lambda x: 'Acidic' if x < 7.0 else ('Basic' if x > 7.0 else 'Neutral'))
df
```

```python
def classify_ph(x):
    if x < 7.0:
        return 'Acidic'
    elif x > 7.0:
        return 'Basic'
    else:
        return 'Neutral'

temp['acid_base'] = temp['ph_level'].apply(classify_ph)
temp
```

```python
# Q3-3. score 컬럼을 만들어주세요.
    # score = temperature_C * ph_level

df['score'] = df['temperature_C'] * df['ph_level']
df
```

## 7. groupby 예제
- Q7-1. researcher별 평균 온도는 얼마입니까?
- Q7-2. researcher와 온도 상태별 실험 수는 어떻게 됩니까?

```python
# Q7-1. researcher별 평균 온도는 얼마입니까?

df.groupby('researcher')['temperature_C'].mean()
```

```python
# Q7-2. researcher와 온도 상태별 실험 수는 어떻게 됩니까?

df.groupby(['researcher', 'temp_status']).size()
```

## 8. 결측값 처리 실습용
- Q8-1. 3번째행의 'ph_level' 값을 빈 값으로 만들어주세요
- Q8-2. 결측값 존재하는 행을 확인해주세요
- Q8-3. 'ph_level'컬럼의 결측값을 평균값으로 채워주세요

```python
# Q8-1. 3번째행의 'ph_level' 값을 빈 값으로 만들어주세요.

df.loc[2, 'ph_level'] = np.nan
df
```

```python
# Q8-2. 결측값 존재하는 행을 확인해주세요

df[df['ph_level'].isna()]
```

```python
# Q8-3. 'ph_level'컬럼의 결측값을 평균값으로 채워주세요

df['ph_level'] = df['ph_level'].fillna(df['ph_level'].mean())
df
```
